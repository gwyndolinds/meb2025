## Estimadores de máxima verosimilitud para la Normal

Sea un conjunto de datos i.i.d. 

$$
\mathcal{D} = \{x_1, \dots, x_N\} \sim \mathcal{N}(\mu, \sigma^2).
$$

La densidad de un dato es

$$
f(x_j;\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} 
\exp\!\left(-\frac{(x_j-\mu)^2}{2\sigma^2}\right).
$$

La verosimilitud es

$$
L(\mu,\sigma^2) = \prod_{j=1}^N f(x_j;\mu,\sigma^2) 
= (2\pi\sigma^2)^{-N/2}
\exp\!\left(-\frac{1}{2\sigma^2}\sum_{j=1}^N (x_j-\mu)^2\right).
$$

La log-verosimilitud es

$$
\ell(\mu,\sigma^2) = \log L(\mu,\sigma^2)
= -\frac{N}{2}\log(2\pi\sigma^2) 
-\frac{1}{2\sigma^2}\sum_{j=1}^N (x_j-\mu)^2.
$$

### Estimador de $\mu$

$$
\frac{\partial \ell}{\partial \mu}
= \frac{1}{\sigma^2}\sum_{j=1}^N (x_j-\mu).
$$

Igualando a cero:

$$
\sum_{j=1}^N (x_j-\mu) = 0 
\quad \Longrightarrow \quad 
\hat{\mu}_{MLE} = \frac{1}{N}\sum_{j=1}^N x_j.
$$

La segunda derivada es

$$
\frac{\partial^2 \ell}{\partial \mu^2} = -\frac{N}{\sigma^2} < 0,
$$


### Estimador de $\sigma^2$

$$
\frac{\partial \ell}{\partial \sigma^2}
= -\frac{N}{2\sigma^2} 
+ \frac{1}{2(\sigma^2)^2}\sum_{j=1}^N (x_j-\mu)^2.
$$

Igualando a cero:

$$
\sum_{j=1}^N (x_j-\mu)^2 = N\sigma^2.
$$
$$
\hat{\sigma}^2_{MLE} 
= \frac{1}{N}\sum_{j=1}^N (x_j-\hat{\mu}_{MLE})^2.
$$

Consideramos un conjunto de datos 

$$
\mathcal{D} = \{x_1, \dots, x_N\}
$$

independientes e idénticamente distribuidos, los cuales suponemos que provienen de una distribución Gamma, con la siguiente pdf

$$
p(x | a) = \frac{a^5}{4!} x^4 e^{-ax}
$$

Encuentre el estimador de máxima verosimilitud para $a$.

$$
\mathcal{D}=\{x_1,x_2,\dots,x_N\},
$$
con densidad
$$
p(x\mid a)=\frac{a^5}{4!},x^4 e^{-ax}
$$
### Vero
$$
L(a)=\prod_{j=1}^N p(x_j\mid a)
= \prod_{j=1}^N \frac{a^5}{4!}\, x_j^4 e^{-a x_j}.
$$
Agrupando factores:
$$
L(a)=\left(\frac{a^5}{4!}\right)^N \left(\prod_{j=1}^N x_j^4\right)
\exp\!\left(-a\sum_{j=1}^N x_j\right).
$$
#### Log-verosimilitud
Tomamos logaritmo:
$$
\ell(a)=\log L(a)
= N\cdot 5\log a - N\log(4!) + 4\sum_{j=1}^N \log x_j - a\sum_{j=1}^N x_j.
$$


### Derivada y condición de máximo

Derivamos $\ell(a)$ respecto a $a$ y la igualamos a cero:
$$
\frac{d\ell}{da} = \frac{5N}{a} - \sum_{j=1}^N x_j = 0.
$$
De aquí se obtiene el estimador de máxima verosimilitud
$$
\hat a_{MLE} = \frac{5N}{\sum_{j=1}^N x_j}.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
#datos generados
x = np.linspace(0, 1, 1001)
y = np.tan(np.e**-1 * np.pi * x) + np.random.normal(scale=0.31, size=x.shape)
```

```python
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#regresion simple
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y,shuffle=True, test_size=0.2, random_state=1)

modelo = LinearRegression()

modelo.fit(X_train[:,None],y_train)
modelo.score(X_train[:,None],y_train), modelo.score(X_test[:,None],y_test)
```

```python
plt.plot(X_train, y_train, 'or', label='Train')
plt.plot(X_test, y_test, 'ob', label='Test')
x_fit = np.linspace(0, 1, 100)
y_fit = modelo.predict(x_fit[:, None])
plt.plot(x_fit, y_fit, '-c', label='Prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
```

```python
modelo_regularizado = Pipeline(steps=[
    ('poly', PolynomialFeatures(degree=10, include_bias=True)),
    ('scaler', StandardScaler()),
    ('ridge', Ridge(alpha=50))
])
```

```python
modelo_regularizado.fit(X_train[:,None],y_train)
modelo_regularizado.score(X_train[:,None],y_train), modelo_regularizado.score(X_test[:,None],y_test)
```

```python
plt.plot(X_train, y_train, color = 'pink', marker='o',linestyle='',label='Train')
plt.plot(X_test, y_test, marker='o', color='magenta', linestyle='', label='Test')

x_fit = np.linspace(0, 1, 100)
y_fit = modelo_regularizado.predict(x_fit[:, None])
plt.plot(x_fit, y_fit, '-c', label='Prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```
